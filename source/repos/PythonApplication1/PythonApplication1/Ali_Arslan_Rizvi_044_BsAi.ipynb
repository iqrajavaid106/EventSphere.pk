## **Task 1**
### **Built-in Functions in Python**
 #### **Introduction :**

Built-in functions in Python are pre-defined functions that are available for immediate use. They simplify common programming tasks such as calculations, data processing, and iteration. In this section, different built-in functions are demonstrated using practical examples, mainly focusing on lists.

#### **Basic Functions :**

##### **1. Print()**

**Description:** Displays output on the screen.

```python
marks = [78, 85, 92, 66, 88]
print("Student Marks:", marks)
```

##### **2. len()**
Description: Returns the number of elements in a list.

```python
marks = [78, 85, 92, 66, 88]
print("Total Students:", len(marks))
```

##### **3. type()**

Description: Returns the data type of a variable.

```python
marks = [78, 85, 92]
print("Type of marks:", type(marks))
```

#### **List-Based Functions :**

##### **4. sum()**

Description: Calculates total of list values.

```python
marks = [78, 85, 92, 66, 88]
print("Total Marks:", sum(marks))
```

##### **5. max()**

Description: Returns highest value.

```python
print(max(marks))
```

##### **6. min()**

Description: Returns lowest value.

```python
print(min(marks))
```

##### **7. sorted()**

Description: Returns sorted list.

```python
print(sorted(marks))
```

##### **8. all()**

Description: Checks if all values are true.

```python
print(all(marks))
```

##### **9. any()**

Description: Checks if any value is true.

```python
print(any(marks))
```

##### **10. enumerate()**

Description: Adds index to elements.

```python
for i, v in enumerate(marks):
    print(i, v)
```

##### **11. zip()**

Description: Combines two lists.

```python
names = ["Ali", "Ahsan", "Moeez"]
print(list(zip(names, marks)))
```

#### **Numeric Functions :**

##### **12. abs()**

The abs() function returns the absolute value of a number, which means it removes any negative sign. It is commonly used in mathematical calculations where only the magnitude of a value matters, such as distance or error calculations.

```python
print(abs(-20))
```

##### **13. round()**

The round() function rounds a number to the nearest integer or to a specified number of decimal places. It is useful when you want to simplify floating-point numbers for display or reporting purposes.

```python
print(round(84.6))
```

```python
print(round(84.655,2))
```

##### **14. pow()**

The pow() function raises a number to a given power. For example, pow(2, 3) calculates 2
3
. It is widely used in mathematical and scientific computations.

```python
print(pow(2, 3))
```

#### **Iteration :**

##### **15. range()**

The range() function generates a sequence of numbers and is mainly used in loops. It helps in repeating a block of code a specific number of times, making it essential for iteration-based tasks.

```python
for i in range(5):
    print(i)
```

#### **Type Conversion :**

##### **16. int()**

The int() function converts a value into an integer. It is commonly used to convert numeric strings or floating-point numbers into whole numbers for calculations.

```python
print(int("10"))
```

##### **17. float()**

The float() function converts a value into a decimal number. It is useful when precision is required in mathematical operations involving non-integer values.

```python
print(float("10"))
```

##### **18. str()**

The str() function converts any data type into a string format. It is especially useful when combining numbers with text for display purposes.

```python
print(str(100))
```

#### **Input & Misc :**

### **Synthetic Population Dataset**
 #### **Objective :**
 To generate a large-scale synthetic dataset representing the population of Sialkot City using Python libraries such as NumPy and random. The dataset includes attributes like names, age, gender, height, and weight.
 #### **Introduction**
 In data science, real-world data is not always available due to privacy, cost, or accessibility issues. Therefore, synthetic data is generated to simulate real populations for analysis and testing. In this task, we create a simulated dataset of Sialkot City’s population using Python. The dataset represents 1.2 million individuals with randomly generated attributes such as names, age, gender, height, and weight. NumPy is used for efficient numerical operations, while the random module helps introduce variability in the data.

##### **Step 1: Import Required Libraries**

**Explaination**

In this step, we import necessary Python libraries. NumPy is used for numerical computations and handling large arrays, while the random module is used to generate random selections such as names.

```python
import numpy as np
import random
```

##### **Step 2: Create Name Pools**

**Explanation:**

We create two lists: one for first names and one for last names. These lists act as a pool from which random names will be generated. This helps in creating realistic-looking human names.

```python
first_names = [
    "Ali", "Ahmed", "Usman", "Hassan", "Bilal",
    "Saad", "Hamza", "Omar", "Zain", "Ibrahim",
    "Ayaan", "Danish", "Faisal", "Imran", "Raza",
    "Ayesha", "Fatima", "Zainab", "Hira", "Sara",
    "Noor", "Maryam", "Hania", "Laiba", "Kiran"
]

last_names = [
    "Khan", "Malik", "Sheikh", "Butt", "Chaudhry",
    "Rana", "Iqbal", "Siddiqui", "Ansari", "Qureshi",
    "Nawaz", "Shah", "Mirza", "Abbasi", "Farooq"
]
```

##### **Step 3: Define Population Size**

**Explanation:**

Here we define the total number of synthetic individuals. We assume a population of 1.2 million to simulate Sialkot City on a large scale.

```python
population_size = 1200000
```

##### **Step 4: Generate Synthetic Names**

**Explanation:**

In this step, we randomly combine first names and last names to create full names. This simulates a realistic population dataset.

```python
names = np.array([
    random.choice(first_names) + " " + random.choice(last_names)
    for _ in range(population_size)
])
```

##### **Step 5: Generate Age Data**
**Explanation:**

We generate random ages between 1 and 80 to represent children, adults, and elderly individuals in the population.

```python
ages = np.random.randint(1, 80, population_size)
```

##### **Step 6: Generate Gender Data**

**Explanation:**

Gender is randomly assigned as Male or Female to simulate real population distribution.

```python
genders = np.random.choice(["Male", "Female"], population_size)
```

##### **Step 7: Generate Height Data**

**Explanation:**

We generate height values using a normal distribution centered around 165 cm, which reflects a realistic human height pattern.

```python
heights = np.random.normal(165, 10, population_size).astype(int)
```

##### **Step 8: Generate Weight Data**

**Explanation:**

Weight is also generated using a normal distribution centered around 65 kg to reflect natural variation in human body weight.



```python
weights = np.random.normal(65, 12, population_size).astype(int)
```

##### **Step 9: Combine All Data into Dataset**

**Explanation:**

All generated attributes (names, age, gender, height, weight) are combined into a single structured dataset using NumPy’s column stacking method.

```python
dataset = np.column_stack((names, ages, genders, heights, weights))
```

##### **Step 10: Display Sample Data**

**Explanation:**

Since the dataset is very large (1.2 million rows), we only display the first 10 records for verification.

```python
print("Dataset Shape:", dataset.shape)
print("\nFirst 10 Records:\n")
print(dataset[:10])
```

##### **Step 11: Summary Statistics**

**Explanation:**

We calculate basic statistics such as average age, height, and weight to analyze the dataset.

```python

print("Total Population:", population_size)
print("Average Age:", np.mean(ages))
print("Average Height:", np.mean(heights))
print("Average Weight:", np.mean(weights))
```
